```
# Import libraries
import numpy as np
import math
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Given data points
x = np.array([1, 2, 2, 3]).reshape(-1, 1)
y = np.array([1, 2, 3, 6])

# Train regression model
model = LinearRegression()
model.fit(x, y)

# Predicted values
y_pred = model.predict(x)

# -------- RMSE using sklearn --------
rmse_sklearn = math.sqrt(mean_squared_error(y, y_pred))
print("Root mean square error (sklearn):", rmse_sklearn)

# -------- RMSE mathematical calculation --------
# residuals = actual - predicted
residuals = y - y_pred

# square of residuals
squared_residuals = residuals ** 2

# mean of squared residuals
mse = np.mean(squared_residuals)

# root of mse
rmse_manual = math.sqrt(mse)

print("Residuals:", residuals)
print("Squared Residuals:", squared_residuals)
print("Mean Squared Error:", mse)
print("Root mean square error (manual):", rmse_manual)
```
